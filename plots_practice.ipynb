```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return  x ** 2 + 6 * x - 5 
```

```python
xs = np.arange(-10, 6, 0.25)
ys = f(xs)

plt.plot(xs, ys)
```

```python
xs = np.arange(-15, 10, 0.25)
ys = f(xs)

def tangent_line(x):
    return slope * (x - x0) + y0

x0 = 3.0
y0 = f(x)

# slope = dy/dx or 
#  slope = (f(x+h) + f(x)) / h

# slope = 2 * x + 6
h = 0.00001
slope = (f(x+h) - f(x)) / h

txs = np.linspace(-1, 10, 20)
tys = tangent_line(txs)

plt.title("Function f(x) and Its Tangent Line")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(xs, ys, label="f(x) = x^2 + 6x - 5 ")
plt.plot(txs, tys, linestyle='dashed', label="tangent line")
plt.scatter(x0, y0, color="black", label="point of tangency")
plt.legend()

```
